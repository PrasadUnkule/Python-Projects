# Outlier Detection
Two Approaches
1. Remove the Rows where we have outliers
2. Handle the outliers

```python
data.info()
```

```python
data.boxplot()
```

```python
df3=data.copy()
```

```python
# 1.5 IQR- IQR Method
# Q3-Q1
IQR=df3['MILES*'].quantile(0.75)-df3['MILES*'].quantile(0.25)
IQR
```

```python
lower=df3['MILES*'].quantile(0.25)-(1.5*IQR)
upper=df3['MILES*'].quantile(0.75)+(1.5*IQR)
```

```python
print(lower)
print(upper)
```

```python
df3[df3['MILES*']<-8.35]
```

```python
df3[df3['MILES*']>21.65]
```

```python
df3[df3['MILES*']>21.65].count()
```

```python
# How to Remove the outlier
df3.drop(df3[df3['MILES*']>21.65].index, inplace=True)
```

```python
df3.shape
```

```python
df3.boxplot()
```

```python
df3.hist()
```

```python
# Handle the Outlier
# If we replace all the values greater than 21.65, there will be no outliers
df4=data.copy()
```

```python
df4.loc[df4['MILES*']>21.65,'MILES*']=21.65
```

```python
df4[df4['MILES*']>21.65].count()
```

```python
df4.boxplot()
```

```python
# 1.5 IQR rule is statistical way to find outlier
# Business change the rule for the quantile 
# If a business says value less than 5 percentile or greater the 95 percentile is outlier, we need to do according to that
lower=df3['MILES*'].quantile(0.05)-(1.5*IQR)
upper=df3['MILES*'].quantile(0.95)+(1.5*IQR)
```

```python
print(lower)
print(upper)
```

# Merge Two Data Frames
# Concatenate or Add Two Data Frames


```python
first=pd.DataFrame({'ID':[1,2,3,4,5], 
                    'Name':['A','B','C','D','E'], 
                    'SubjectID':['sub1','sub2','sub3','sub4','sub5']})
```

```python
first
```

```python
second=pd.DataFrame({'ID':[1,2,3,4,5], 
                    'Name':['X','Y','Z','P','Q'], 
                    'SubjectID':['sub4','sub8','sub9','sub2','sub5']})
```

```python
second
```

```python
new=pd.merge(first,second,on='ID')
```

```python
new
```

```python
new1=pd.merge(first,second,on=['ID','SubjectID'])
```

```python
new1
```

# Types of merge
1. Left
2. Right
3. Inner
4. Outer

```python
new1=pd.merge(first,second,on='SubjectID', how='left')
```

```python
new1
```

```python
new1=pd.merge(first,second,on='SubjectID', how='right')
new1
```

```python
new1=pd.merge(first,second,on='SubjectID', how='inner')
new1
```

```python
new1=pd.merge(first,second,on='SubjectID', how='outer')
new1
```

```python
# Concatenate the dataframe
pd.concat([first,second])
```
